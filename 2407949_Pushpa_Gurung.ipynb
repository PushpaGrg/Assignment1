Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
#Calculate first quartile and third quartile based on 'score'
Q1_score=south_asian_data['score'].quantile(0.25) #1at quartile
Q3_score=south_asian_data['score'].quantile(0.75) #3rd quartile
#Calculate interquartile range for 'score'
IQR_score=Q3_score-Q1_score
#Find the lower bound and upper bound for outliers in 'score' using 1.5*IQR rule
lower_bound_score=Q1_score-1.5*IQR_score
upper_bound_score=Q3_score+1.5*IQR_score
#Calculate first quartile and second quatile based on 'Log GDP per capita'
Q1_gdp=south_asian_data['Log GDP per capita'].quantile(0.25) #1st quartile
Q3_gdp=south_asian_data['Log GDP per capita'].quantile(0.75) #3rd quartile
#Calculate interquartile range for 'Log GDP per capita'
IQR_gdp=Q3_gdp-Q1_gdp
#Find the lower bound and upper bound for outliers in 'Log GDP per capita' using 1.5*IQR rule
lower_bound_gdp=Q1_gdp-1.5*IQR_gdp
upper_bound_gdp=Q3_gdp+1.5*Q1_gdp
```

```python
#Create a new column 'Outliers' to flag rows as outliers
south_asian_data['Outliers']=(south_asian_data['score']<lower_bound_score)|(south_asian_data['score']>upper_bound_score)|(south_asian_data['Log GDP per capita']<lower_bound_gdp)|(south_asian_data['Log GDP per capita']>upper_bound_gdp)
```

```python
#Filter outliers and non-outliers
outliers=south_asian_data[south_asian_data['Outliers']]
non_outliers=south_asian_data[~south_asian_data['Outliers']]
```

```python
#Plot Outliers and non-outliers
sns.scatterplot(data=outliers,x='Log GDP per capita',y='score',color='red',s=100,marker='x')
sns.scatterplot(data=non_outliers,x='Log GDP per capita',y='score',color='blue',s=100,marker='o')
#Add title of the plot
plt.title('Outliers in GDP per Capita vs Score in South Asian Countries',weight='bold')
plt.xlabel('GDP per Capita') #set x-axis label
plt.ylabel('Happiness Score') #set y-axis label
#Add legend and manually set the labels
plt.legend(labels=['Outliers','Non-Outliers'])
#Show plot
plt.show()
```

According to the above scatterplot, there is only one outlier which is denoted by red 'X' mark.The plot indicates that this outlier country has a very low GDP per capita around 0.6 and also a low happiness score. This often represent poor quality of life, political instability, economic instability and so on.
Due to this outlier, there may be reduction in regional average for both metrics(GDP and Score). Also there can be distortion in relationship between the metrics .

```python
#Select the top 3 positive and negative GDP-Score Gap
top3_positive=descending_rank.head(3)
top3_negative=ascending_rank.head(3)
#Combine the data and strore it into a single dataframe
top_countries=pd.concat([top3_positive,top3_negative])
#Set the plot's size to 10X6 inches
plt.figure(figsize=(10,6))
#Plot the bar chart to visualize GDP-Score Gap for selected countries
sns.barplot(data=top_countries,x='Country name',y='GDP-Score Gap')
#Set the title of plot
plt.title('Top 3 Countries with the Largest and Smallest GDP-Score Gap',weight='bold')
plt.xlabel('Country')
plt.ylabel('GDP-Score Gap')
#Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
#use for loop to label the bars with gap values
for index,value in enumerate(top_countries['GDP-Score Gap']):
  plt.text(index,value,f'{value:.2f}',ha='center',va='bottom' if value>0 else 'top')
plt.show()
```
